Importing Brazilian json

```python
import json
from urllib.request import urlopen
```

```python
with urlopen('https://raw.githubusercontent.com/codeforamerica/click_that_hood/master/public/data/brazil-states.geojson') as response:
 Brazil = json.load(response) # Javascrip object notation 
```

Get rid of accents.

```python
ACENTOS = { 'Á':'A','Ã':'A','À':'A','Â':'A', 
            'Ó':'O','Õ':'O','Ô':'O', 'Ò':'O',
            'É':'E','Ê':'E','È':'E',
            'Í':'I',
            'Ú':'U','Ù':'U',
            'Ç':'C',
}

def tirar_acentos(message):
    cipher = ''
    for letter in message.upper():
        if letter in ACENTOS:
            cipher += ACENTOS[letter]
        else:
            cipher += letter
    return cipher
```

```python
tirar_acentos('Piauí')
```

Dictionary State:ID

```python
state_id_map = {}
for feature in Brazil['features']:
    feature['id'] = feature['properties']['sigla']
    state_id_map[tirar_acentos(feature['properties']['name'].upper())] = feature['id']
```

```python
state_id_map
```

State ID in gas Table

```python
df.head()
```

```python
df['SIGLA']=df['ESTADO'].apply(lambda x: state_id_map[x])
```

Creating the Map of Comum Gas 2004-2019

```python
dfGASCOMUM = df[df['PRODUTO']=='GASOLINA COMUM']
```

Price Mean by Year and State

```python
dfGASCOMUM_ESTADO_ANO = dfGASCOMUM.groupby(['SIGLA', 'ANO']).mean().round(3)
dfGASCOMUM_ESTADO_ANO.reset_index(inplace=True)
```

```python
fig = px.choropleth_mapbox(dfGASCOMUM_ESTADO_ANO, locations = 'SIGLA',
                geojson = Brazil,
                color = 'PREÇO MÉDIO REVENDA',
                animation_frame = 'ANO',
                color_continuous_scale=px.colors.sequential.Inferno_r,
                title="Comum Gas Price 2004 - 2019",
                mapbox_style = "carto-positron",
                center={"lat":-14, "lon": -55},
                hover_data=["PREÇO MÉDIO REVENDA"],
                zoom = 3)
fig.update_geos(fitbounds="locations", visible=False)
```
